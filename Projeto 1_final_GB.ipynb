## Pergunta:

 Será que o pacto militar da OTAN causou algum impacto na expectativa de vida e na mortalidade infantil?

## Hipótese:

 Países que se juntam a OTAN têm um aumento na expectativa de vida, e diminuição na taxa de mortalidade infantil, devido ao pacto militar, pois com ele haveria menos chances de ocorres posíveis guerras ou algo que possa impactar de forma negativa na população.

Em meu projeto pretendo comparar países que pertencem a OTAN e aos que não pertencem, nele analisarei dados como mortalidade infantil, expectativa de vida,( anos de 1980,2000,2015), também montarei tabelas, histogramas, e gráficos comparativos com os dados do países que estão e que não, um ao lado do outro com graficos permitindo-me ter uma visao mais ampla para comparar.
```python
exp = ExpV.set_index('country')
#P_exp.head()
```

```python
mi = MI.set_index('country')
#P_mi.head()
```

Adicionando os anos escolhidos:

```python
ExpV_ = exp.loc[:, ['country',1990,2000,2013]]
MI_ = mi.loc[:, ["country", 1990,2000,2013]]
ExpV_.head()
```

Adicionando os joins:

```python
df = MI_.join(ExpV_, how='inner', lsuffix='_Mort', rsuffix='_Exp')
```

```python
df.head()
```

```python
OTAN= ('Albania', 'Belgium', 'Bulgaria', 'Canada', 'Croatia', 'Czech Republic', 'Denmark', 'Estonia', 'France', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Montenegro', 'Netherlands', 'Norway', 'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Turkey', 'United Kingdom', 'United States')
df['Pertencimento']='Não'
for i in OTAN:
    df.Pertencimento[i]='Sim'
df.head()
```

```python
Df=df.loc[:, ['1990_Exp','2000_Exp','2013_Exp','1990_Mort','2000_Mort','2013_Mort','Pertencimento']]
Df.head()
```

Criando DataFrames diferentes sendo um deles apenas os paise que pertencem a OTAN e outro apenas aos países que não pertencem

```python
Otan = Df.loc[Df.Pertencimento == "Sim"]
N_Otan = Df.loc[Df.Pertencimento == "Não"]
```

# Comparando expectativa de vida:

### Comparando a expectativa de vida dos paises que são e que não são da OTAN nos anos de 1990, 2000, e 2013, e fazendo histogramas para que possam ser melhor analisados:

```python
def verticalLine(plot, x, y, color, nome):
    plot.axvline(x, color=color, linestyle="dashed", linewidth=2)
    plot.text(1.01*x, y, nome, color=color)
```

```python
faixa= np.arange(65,83,1)
```

## Fazendo gráficos comparativos com as médias: Expectativa de vida

```python
media_Exp_Otan_1990= Otan['1990_Exp'].mean()
media_Exp_N_Otan_1990= N_Otan['1990_Exp'].mean()


media_Exp_Otan_2000= Otan['2000_Exp'].mean()
media_Exp_N_Otan_2000= N_Otan['2000_Exp'].mean()


media_Exp_Otan_2013= Otan['2013_Exp'].mean()
media_Exp_N_Otan_2013= N_Otan['2013_Exp'].mean()


media_Exp_Otan = (media_Exp_Otan_1990+media_Exp_Otan_2000+media_Exp_Otan_2013)/3
media_Exp_N_Otan = (media_Exp_N_Otan_1990+media_Exp_N_Otan_2000+media_Exp_N_Otan_2013)/3


medias_Exp = {'OTAN': media_Exp_Otan , 'Países fora da OTAN': media_Exp_N_Otan}
names=list(medias_Exp.keys())
values=list(medias_Exp.values())

plt.bar(names,values, color = ('r','b'))
plt.title("Expectativa de vida (Médias)")
plt.ylabel('Expectativa de Vida ')
plt.show()
```

Este é um grafico de barras que nos mostra as médias dos países que estão dentro e fora da OTAN. Nele percebemos que a expectativa de vida dos países membros da OTAN é maior do que a dos países que estam fora.

```python
anos= [1990,2000,2013]
medias_Exp_Otan =[media_Exp_Otan_1990,media_Exp_Otan_2000,media_Exp_Otan_2013]
medias_Exp_N_Otan =[media_Exp_N_Otan_1990,media_Exp_N_Otan_2000,media_Exp_N_Otan_2013]

plt.scatter(anos,medias_Exp_Otan, label = 'OTAN', color= 'r')
a = np.polyfit(anos,medias_Exp_Otan, 1)
b = np.poly1d(a)
plt.plot(anos,b(anos),"r--")


plt.scatter(anos,medias_Exp_N_Otan, label = 'Países fora da OTAN', color= 'b')
a = np.polyfit(anos,medias_Exp_N_Otan, 1)
b = np.poly1d(a)
plt.plot(anos,b(anos),"b--")

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel('Expectativa de vida')
plt.xlabel('Anos')
plt.title('Expectativa de vida')
plt.show()

print("A média da expectativa de vida dos países membros da OTAN foi de {}",(media_Exp_Otan))
print("A média da expectativa de vida dos países não membros da OTAN foi de {}",(media_Exp_N_Otan))


```

Este gráfico ("scatter") nos mostra a média da expectativa de vida ao longo dos anos, ele também compara os países membros da OTAN e os não membros. Analisando ele percebemos que a expectiva de vida dos membros da OTAN e dos não membros crescem de uma maneira muito parecida, mas é bem perceptível que os membros da OTAN tem uma expectativa de vida muito maior do que os que estam fora.

## Fazendo gráficos comparativos com as médias: Mortalidade infantil

```python
media_Mort_Otan_1990= Otan['1990_Mort'].mean()
media_Mort_N_Otan_1990= N_Otan['1990_Mort'].mean()


media_Mort_Otan_2000= Otan['2000_Mort'].mean()
media_Mort_N_Otan_2000= N_Otan['2000_Mort'].mean()


media_Mort_Otan_2013= Otan['2013_Mort'].mean()
media_Mort_N_Otan_2013= N_Otan['2013_Mort'].mean()


media_Mort_Otan = (media_Mort_Otan_1990+media_Mort_Otan_2000+media_Mort_Otan_2013)/3
media_Mort_N_Otan = (media_Mort_N_Otan_1990+media_Mort_N_Otan_2000+media_Mort_N_Otan_2013)/3


medias_Mort = {'OTAN': media_Mort_Otan , 'Países fora da OTAN': media_Mort_N_Otan}
names=list(medias_Mort.keys())
values=list(medias_Mort.values())

plt.bar(names,values, color = ('r','b'))
plt.title("Mortalidade infantil (Médias)")
plt.ylabel('Mortalidade infantil ')
plt.show()

print("A média de mortalidade infantil dos países membros da OTAN foi de {}",(media_Mort_Otan))
print("A média de mortalidade infantil dos países não membros da OTAN foi de {}",(media_Mort_N_Otan))
```

Este gráfico sobre mortalidade infantil, em que divide os membros da OTAN e os de fora da OTAn deixa bem clara uma difernça de valores, enquanto em um lado do gráfico onde os membros da OTAN tem uma média de aproximadamente 9,5 mortes a cada 1000 nascimentos, a dos países não membros da OTAN tem uma média absurdamente maior com um valor aproximado de 61 mortes a cada 1000 nascimentos.

```python
anos= [1990,2000,2013]
medias_Mort_Otan =[media_Mort_Otan_1990,media_Mort_Otan_2000,media_Mort_Otan_2013]
medias_Mort_N_Otan =[media_Mort_N_Otan_1990,media_Mort_N_Otan_2000,media_Mort_N_Otan_2013]

plt.scatter(anos,medias_Mort_Otan, label = 'OTAN', color= 'r')
a = np.polyfit(anos,medias_Mort_Otan, 1)
b = np.poly1d(a)
plt.plot(anos,b(anos),"r--")


plt.scatter(anos,medias_Mort_N_Otan, label = 'Países fora da OTAN', color= 'b')
a = np.polyfit(anos,medias_Mort_N_Otan, 1)
b = np.poly1d(a)
plt.plot(anos,b(anos),"b--")

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel('Mortalidade infantil')
plt.xlabel('Anos')
plt.title('Mortalidade infantil')
plt.show()
```
